```python
from math import floor

x, y = map(int, input().split())



start = 1
end = x

def solution(start, end):
    z = floor((100 * y) / x)
    
    if z >= 99:
            return -1
        
    while start <= end:

        mid = (start + end) // 2
        
        nx = x + mid
        ny = y + mid
        
        if floor((100 * ny) / nx) <= z:
            start = mid + 1
        else:
            end = mid - 1
    return start
            
answer = solution(start, end)
print(answer)
```
